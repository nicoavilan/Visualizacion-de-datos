# Uso de GeoPandas

GeoPandas: Los GeoDataFrames tienen una columna "Geometry" que se expresa según tres tipos de objetos:
* Puntos.
* Líneas.
* Polígono

```python
import pandas as pd
import matplotlib.pyplot as plt
import geopandas
import warnings
warnings.filterwarnings("ignore")
```

Datos disponibles

```python
world = geopandas.read_file(geopandas.datasets.get_path("naturalearth_lowres"))
world
```

```python
world.plot(figsize= (10,7))
plt.show()
```

Eventos sísmicos

```python
url="https://raw.githubusercontent.com/nicoavilan/Bases-de-datos/main/Earthquake.csv"
df=pd.read_csv(url)
df
```

```python
df["date_time"] = pd.to_datetime(df["Date"] +" "+ df["Time"], format = "%m/%d/%Y %H:%M:%S", errors = "coerce")
```

```python
df.loc[[3377, 3378, 7512, 20650]]
```

```python
df = df.drop([3378, 7512, 20650])
```

```python
df["date_time"] = pd.to_datetime(df["Date"] +" "+ df["Time"], format = "%m/%d/%Y %H:%M:%S")
```

```python
df_ = df[['date_time','Magnitude', 'Latitude', 'Longitude']].reset_index(drop = True)
df_
```

```python
gdf = geopandas.GeoDataFrame(df_, geometry = geopandas.points_from_xy(df_.Longitude, df_.Latitude ))
gdf.head(2)
```

```python
gdf.geometry.plot();
```

```python
ax = world.plot(color= "green", edgecolor= "black", figsize=(8,6))
gdf.geometry.plot(ax= ax, color= "red", markersize= 3, alpha= 0.3, label="Eventos\nSísmicos")
plt.title("Ubicación de Eventos Sísmicos", size= 15)
plt.legend()
plt.show()
```

Terremoto de Japón

```python
gdf[ gdf.Magnitude > 9 ]
```

```python
ax = world.plot(color= "green", edgecolor= "black", figsize=(8,6))
gdf[ gdf.Magnitude > 9 ].geometry.plot(ax= ax, color= "r", marker = "*",
                                       markersize= 200, alpha= 1, label="Eventos\nSísmicos\nFuertes")
plt.title("Ubicación de Eventos Sísmicos", size= 15)
plt.legend()
plt.show()
```

```python
gdf_japan = gdf.query("25 < Latitude < 50 and 125 < Longitude < 150") # Japón
gdf_japan
```

```python
ax = world.plot(color= "green", edgecolor= "black", figsize=(8,6))
gdf_japan.geometry.plot(ax= ax, color= "r", 
                                       markersize= 2, alpha= 1, label="Japón")
gdf[gdf.Magnitude > 9].geometry.plot(ax= ax, color= "r", marker = "*",
                                       markersize= 200, alpha= 1, label="Eventos\nSísmicos\nFuertes")
plt.title("Ubicación de Eventos Sísmicos", size= 15)
plt.legend()
plt.show()
```

```python
ax = world[world.name == "Japan"].plot(color= "white", edgecolor= "black", figsize=(7,5))
gdf_japan.geometry.plot(ax= ax, color= "r", markersize= 2, alpha= 1, label="Japón")

plt.show()
```

```python
gdf_japan_T[ gdf_japan_T.Magnitude > 9]
```

```python
#fig= plt.figure()
ax = world[world.name == "Japan"].plot(color= "forestgreen", edgecolor= "black")#, figsize=(7,5))#, color= "LightGreen")#, ocean_color= "blue")
gdf_japan_T.geometry.plot(ax= ax, color= "orange", markersize= 3, alpha= 1, label="Réplicas", edgecolor="k")
gdf_japan_T.loc[[1]].geometry.plot(ax= ax, color= "r", marker = "*",
                                       markersize= 150, alpha= 1, label="Epicentro")
plt.xlabel("Longitud")
plt.ylabel("Latitud")
plt.title("Réplicas del Terremoto de Japón - 2011\n", weight= "bold", size = 15)
#ax.add_feature(cartopy.feature.OCEAN, facecolor=ocean_color, edgecolor='none')
plt.legend()
plt.show()
```

```python
T = gdf_japan_T["date_time"]
```

```python
T[3].date()
```

```python
x = gdf_japan_T.Longitude.values
y = gdf_japan_T.Latitude.values
```

```python
import matplotlib.animation as animation
from IPython.display import HTML
```

```python
fig= plt.figure()
ax=fig.gca()

def actualizar(i):
  ax.clear()
  world[world.name == "Japan"].plot(ax = ax, color= "forestgreen")
  ax.set_title("Réplicas del Terremoto de Japón - 2011\n", weight= "bold", size = 15)
  ax.text(130 , 44, f"Fecha\n{T[i].date()}", size = 12, weight= "bold")
  ax.text(130 , 42, f"Hora\n{T[i].time()}", size = 12, weight= "bold")
  ax.set_xlabel("Longitud")
  ax.set_ylabel("Latitud")
  gdf_japan_T.loc[[1]].geometry.plot(ax= ax, color= "orange", marker = "*",
                                       markersize= 150, alpha= 1, label="Epicentro")
  ax.plot(x[:i],  y[:i],'o',markersize=2,color='r', alpha = 1, label= "Réplicas")
  ax.axis(xmin= 125, xmax= 148, ymin= 30, ymax= 50)
  ax.legend()
plt.show()
```

```python
ani=animation.FuncAnimation(fig, actualizar, range(0,len(x)), interval= 200)
HTML(ani.to_html5_video())
```
