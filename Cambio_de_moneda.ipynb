# **Datos históricos de cambio de moneda**

```python
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime
import matplotlib.animation as animation
from IPython.display import HTML
```

## Varoles históricos del precio del Euro en Dólares

```python
url = "https://raw.githubusercontent.com/nicoavilan/Bases-de-datos/main/EUR-USD.csv"
df = pd.read_csv(url)
df
```

Hay que cambiar el formato de las fechas

```python
df.loc[0,"Date"]
```

```python
df["Date"] = pd.to_datetime(df["Date"], format = "%Y-%m-%d")
```

```python
df.loc[0,"Date"]
```

```python
plt.figure(figsize= (15,5))
plt.plot(df.Date,df.Open)
plt.title("Precio del Euro en dólares", fontweight="bold", size = 20)
plt.ylabel("Dólares", fontweight="bold", size = 15)
plt.show()
```

```python
x = df["Date"].values
y = df["Open"].values
N = len(df)
```

```python
fig=plt.figure()
ax=fig.gca()

def actualizar(i):
  ax.clear()

  ax.plot(x[:i], y[:i],'-')
  ax.set_title("Precio del Euro en dólares", fontweight="bold", size = 20)
  ax.set_ylabel("Dólares", fontweight="bold", size = 15)
  ax.plot(x[i],  y[i],'o',markersize=10,color='r')
```

```python
ani=animation.FuncAnimation(fig, actualizar, range(0,N,20), interval=150)
HTML(ani.to_html5_video())
```

La animación se observa al ejecutar el código, GitHub no la genera automáticamente.
