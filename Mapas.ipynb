# **Mapas**

Hay que instalar la librería [geopandas](https://geopandas.org/) para manipular información geográfica.

```python
!pip install geopandas
import geopandas
import matplotlib.pyplot as plt
```

##  Mapa de Colombia

```python
url='https://gist.githubusercontent.com/john-guerra/43c7656821069d00dcbc/raw/3aadedf47badbdac823b00dbe259f6bc6d9e1899/colombia.geo.json'
col = geopandas.read_file(url)
col
```

```python
col.plot(color= "white", figsize=(12,7),edgecolor="black")
plt.title("Departametos de Colombia", size=20)
plt.show()
```

```python
col.plot(column = "AREA", figsize=(12,7),edgecolor="black", legend=True)
plt.title("Área - Departametos de Colombia", size=20)
plt.show()
```
